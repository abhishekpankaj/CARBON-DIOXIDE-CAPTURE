# Splitting the dataset into training and testing dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df2.drop(['xCO2'],axis=1),df2['xCO2'], test_size = 0.2,random_state=50)
```

# Feature scaling the input dataset

```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc=StandardScaler()
sc.fit(X_train)
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

# Importing Linear Regression to predict 

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

**Predicted Value**

```python
a=pd.DataFrame(y_pred)
a
```

**Actual Data**

**Importing accuracy to find accuracy of model**

```python
from sklearn.metrics import r2_score
```

```python
y=r2_score(y_test,y_pred)
y
```
